### Why Visualize Data? ü§î

Data visualization is the graphical representation of information and data. By using visual elements like charts, graphs, and maps, data visualization tools provide an accessible way to see and understand trends, outliers, and patterns in data.

Here's why it's so crucial:

  * **Understanding Complex Data:** Raw data, especially large datasets, can be overwhelming. Visualizations simplify complex data, making it easier to grasp insights that might be hidden in tables of numbers.
  * **Identifying Patterns and Trends:** Visuals excel at revealing patterns, trends, correlations, and anomalies that might be invisible in raw data. For example, a scatter plot can quickly show a correlation, or a line chart can highlight a trend over time.
  * **Communicating Insights:** Visualizations are a powerful communication tool. They help convey your findings to a broad audience, even those without a technical background, making data-driven decisions more accessible.
  * **Faster Decision Making:** When insights are presented clearly and concisely through visuals, stakeholders can make faster and more informed decisions.
  * **Storytelling:** Effective visualizations can tell a compelling story about your data, guiding the viewer through the analysis and highlighting key takeaways.
  * **Exploratory Data Analysis (EDA):** Visualizations are essential during the initial stages of data analysis to explore the data, validate assumptions, and formulate hypotheses.



### What is Matplotlib? (Basic Plotting Library, Its Components) üñºÔ∏è

**Matplotlib** is a comprehensive library for creating static, animated, and interactive visualizations in Python. It's the most widely used plotting library in the Python scientific computing community and serves as the foundation for many other plotting libraries (like Seaborn).

It aims to make it easy to create simple plots and challenging to create complex plots, but gives you full control over every aspect of your plot.

**Key Components of a Matplotlib Plot:**

Understanding these components is crucial, especially when using the object-oriented approach:

1.  **Figure:** The entire window or page that contains everything. Think of it as the canvas on which you draw. A Figure can contain multiple Axes.
      * `plt.figure()` creates a Figure.
2.  **Axes (plural of Axis):** This is the actual plot area where the data is plotted. A Figure can have multiple Axes, each representing a distinct plot. Each Axes object has an x-axis, a y-axis, a title, x-label, y-label, etc.
      * `fig.add_subplot()` or `fig.add_axes()` (for object-oriented) or `plt.subplot()`/`plt.subplots()` (for `pyplot`) create Axes.
3.  **Axis (singular):** These are the number lines of the plot, controlling the limits of the data and generating ticks (the marks on the axis) and tick labels (the text labels for ticks).
4.  **Artist:** Everything visible on the Figure is an Artist (e.g., Text objects, Line2D objects, collections, etc.). The Figure, Axes, and Axis are all Artists.

Think of it like this: You have a **Figure** (the paper) on which you draw one or more **Axes** (the individual graphs). Each **Axes** has its own X and Y **Axis** (the rulers for each graph). And all the data points, lines, labels, and titles are **Artists** drawn on these Axes and the Figure.

# 5.2 Matplotlib Fundamentals - `pyplot` module

The `pyplot` module (`matplotlib.pyplot`) is a collection of functions that make Matplotlib work like MATLAB. Each `pyplot` function makes some change to a figure, e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

It's the easiest way to get started with plotting in Matplotlib.


```python
import matplotlib.pyplot as plt
import numpy as np

# Set a style for aesthetic plots (optional, but good practice)
plt.style.use('ggplot')
```


### Basic Plotting: `plot()`, `show()` ‚úçÔ∏è

The `plt.plot()` function is the most versatile function in `pyplot`. It takes arrays of values as input and draws lines or markers.

  * **`plt.plot(x, y)`:** Plots `y` versus `x`.
  * **`plt.show()`:** Displays all open figures. Without `show()`, plots might not appear depending on your environment (e.g., script vs. Jupyter notebook).

<!-- end list -->


```python
print("--- Basic Plotting: plot() and show() ---")

# 1. Plotting a simple line graph
x_coords = np.array([0, 1, 2, 3, 4, 5])
y_coords = np.array([0, 1, 4, 9, 16, 25]) # y = x^2

plt.plot(x_coords, y_coords)
plt.title("Simple Line Plot of Y = X^2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True) # Add a grid for readability
plt.show() # Display the plot

# 2. Plotting multiple lines on the same graph
x_vals = np.linspace(0, 10, 100) # 100 points between 0 and 10
y1_vals = np.sin(x_vals)
y2_vals = np.cos(x_vals)

plt.plot(x_vals, y1_vals)
plt.plot(x_vals, y2_vals)
plt.title("Sine and Cosine Waves")
plt.xlabel("Angle (radians)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show() # Display the plot
```


**Code Explanation:**

  * `plt.plot(x, y)`: Connects points `(x[i], y[i])` with lines.
  * `plt.title()`, `plt.xlabel()`, `plt.ylabel()`: Set the title and axis labels.
  * `plt.grid(True)`: Adds a grid to the plot, making it easier to read values.
  * `plt.show()`: Renders and displays the plot. If you're in a script, this will open a new window with the plot. In Jupyter/IPython, it renders inline.



### Figure and Axes Objects (Object-Oriented Approach vs. `pyplot` Interface) üìê

Matplotlib offers two main ways to interact with plots:

1.  **`pyplot` interface (explicitly above):** This is the simpler, MATLAB-like interface. Functions like `plt.plot()`, `plt.title()` automatically operate on the "current" Figure and Axes. Good for quick plots.
2.  **Object-Oriented (OO) interface:** This is generally recommended for more complex plots or when you need fine-grained control. You explicitly create `Figure` and `Axes` objects and then call methods on those objects.

<!-- end list -->


```python
print("\n--- Figure and Axes Objects (Object-Oriented Approach) ---")

# Create a Figure and a single Axes object
fig, ax = plt.subplots(figsize=(8, 5)) # Creates a figure and a set of subplots (1x1 by default)

# Plot data on the Axes object
x_data = np.arange(0, 10, 0.1)
y_data = x_data**2

ax.plot(x_data, y_data, label='Y = X^2') # Use ax.plot() instead of plt.plot()

# Set properties using Axes object methods
ax.set_title("Object-Oriented Plot of Y = X^2")
ax.set_xlabel("X-axis (OO)")
ax.set_ylabel("Y-axis (OO)")
ax.grid(True)
ax.legend() # Show the legend

plt.show() # Display the plot
```


**Code Explanation:**

  * `fig, ax = plt.subplots()`: This is the most common way to start an OO plot. It creates a `Figure` object (`fig`) and one or more `Axes` objects (`ax`). By default, it creates a 1x1 grid of subplots, so `ax` is a single `Axes` object.
  * `ax.plot()`: Now we call methods on the `ax` object (e.g., `ax.plot()`, `ax.set_title()`, `ax.set_xlabel()`). This gives explicit control over which plot element you are modifying.

**Why Object-Oriented?**

  * **Clarity:** It's clearer which plot you're modifying when you have multiple plots.
  * **Control:** Gives you access to the full API of Figure and Axes objects, enabling highly customized plots.
  * **Multiple Plots:** Essential when creating complex layouts with multiple subplots.

### Adding Titles, Labels, and Legends üìù

These elements are crucial for making your plots understandable.


```python
print("\n--- Adding Titles, Labels, and Legends ---")

years = np.array([2018, 2019, 2020, 2021, 2022, 2023, 2024])
sales_a = np.array([100, 120, 110, 130, 150, 140, 160])
sales_b = np.array([80, 90, 105, 115, 120, 135, 150])

plt.figure(figsize=(10, 6)) # Create a new figure with specific size

plt.plot(years, sales_a, label='Company A Sales')
plt.plot(years, sales_b, label='Company B Sales')

# Title for the entire plot
plt.title("Annual Sales Performance (2018-2024)", fontsize=16, color='darkblue')

# X and Y axis labels
plt.xlabel("Year", fontsize=12, color='gray')
plt.ylabel("Sales (in thousands $)", fontsize=12, color='gray')

# Add a legend to differentiate the lines
plt.legend(loc='upper left', fontsize=10, shadow=True, fancybox=True)

plt.grid(True, linestyle='--', alpha=0.9) # Customize grid
plt.show()
```


**Code Explanation:**

  * `plt.figure(figsize=(width, height))`: Creates a new figure and sets its size in inches.
  * `plt.title(title_text, fontsize, color)`: Sets the title. You can customize font size, color, etc.
  * `plt.xlabel(label_text, fontsize, color)` / `plt.ylabel(label_text, fontsize, color)`: Set axis labels with similar customization options.
  * `plt.legend()`: Displays the legend.
      * `label`: The text for each line in the legend is specified in the `plt.plot()` call (e.g., `label='Company A Sales'`).
      * `loc`: Controls the legend's position (e.g., 'upper left', 'best', 'lower right').
      * `shadow`, `fancybox`: Aesthetic options for the legend box.

### Customizing Line Styles, Markers, Colors üåà

Matplotlib provides extensive options to customize the appearance of your plots.


```python
print("\n--- Customizing Line Styles, Markers, Colors ---")

x = np.linspace(0, 2 * np.pi, 50)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan_approx = np.tan(x/2) # Just for a third line, ignore math precision

plt.figure(figsize=(12, 7))

# Line 1: Solid blue line with circle markers
plt.plot(x, y_sin, color='blue', linestyle='-', marker='o', label='Sine Wave')

# Line 2: Dashed red line with square markers
plt.plot(x, y_cos, color='red', linestyle='--', marker='s', label='Cosine Wave')

# Line 3: Green dotted line with triangle_up markers, larger size
plt.plot(x, y_tan_approx, color='green', linestyle=':', marker='^', markersize=8, linewidth=2, label='Tangent Approx')

# Shorthand for color, marker, linestyle (e.g., 'ro--' for red circle dashed)
# plt.plot(x, y_sin, 'bo-', label='Sine Wave (Shorthand)')

plt.title("Customized Line Plot of Trigonometric Functions")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()

# Example: Scatter Plot with custom markers and colors
temperatures = np.array([10, 12, 15, 13, 18, 20, 22, 25, 23, 20, 18, 15])
humidity = np.array([70, 65, 60, 68, 55, 50, 45, 40, 48, 52, 58, 62])
cities = ['NY', 'LA', 'SF', 'CH', 'MI', 'AU', 'DA', 'HO', 'SD', 'PH', 'SL', 'BO']

plt.figure(figsize=(10, 6))
# s: size of marker, c: color, alpha: transparency
plt.scatter(temperatures, humidity, s=100, c=temperatures, cmap='viridis', alpha=0.8, edgecolors='black')

# Add city labels next to points (optional, for demonstration)
for i, city in enumerate(cities):
    plt.text(temperatures[i] + 0.5, humidity[i], city, fontsize=9, color='darkgray')

plt.title("Temperature vs. Humidity in Various Cities")
plt.xlabel("Temperature (¬∞C)")
plt.ylabel("Humidity (%)")
plt.colorbar(label='Temperature (¬∞C)') # Add a color bar for the 'c' argument
plt.grid(True, linestyle=':', alpha=0.8)
plt.show()
```


**Common Customization Options for `plt.plot()` and `ax.plot()`:**

  * **`color` (or `c`):** Sets the color of the line/markers. Can be a name ('red', 'blue', 'green'), hex code ('\#FF0000'), or RGB tuple `(R, G, B)`.
  * **`linestyle` (or `ls`):** Sets the style of the line.
      * `-` (solid, default)
      * `--` (dashed)
      * `:` (dotted)
      * `-.` (dash-dot)
  * **`marker`:** Sets the style of the data points.
      * `o` (circle)
      * `s` (square)
      * `^` (triangle up)
      * `x` (x marker)
      * `*` (star)
      * `+` (plus)
      * `D` (diamond)
  * **`linewidth` (or `lw`):** Sets the width of the line.
  * **`markersize` (or `ms`):** Sets the size of the markers.
  * **`alpha`:** Sets the transparency (0.0 = fully transparent, 1.0 = fully opaque).

**Shorthand Format Strings:**
You can combine color, marker, and linestyle into a single string: `'[color][marker][linestyle]'`.
Example: `'ro--'` means red circle dashed line.

**For `plt.scatter()`:**

  * `s`: Size of the markers.
  * `c`: Color of the markers. Can be a single color or an array of values to map to a colormap.
  * `cmap`: Colormap to use if `c` is an array (e.g., 'viridis', 'plasma', 'cividis').
  * `edgecolors`: Color of the marker edges.


### Line Plots: Single and Multiple Lines üìà

We've already seen basic line plots. They are ideal for showing trends over continuous data, especially time series.


```python
print("--- Line Plots ---")

# Data for line plot
time = np.arange(0, 10, 0.1) # 0 to 10 with 0.1 step
temperature = 20 + 5 * np.sin(time) + np.random.normal(0, 0.5, len(time))
humidity = 60 - 3 * np.cos(time) + np.random.normal(0, 0.7, len(time))

plt.figure(figsize=(10, 5))
plt.plot(time, temperature, label='Temperature (¬∞C)', color='red', linestyle='-', marker='o', markersize=4)
plt.plot(time, humidity, label='Humidity (%)', color='blue', linestyle='--', marker='x', markersize=4)

plt.title('Temperature and Humidity over Time')
plt.xlabel('Time (hours)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```


**Key Use Cases:** Trends over time, showing continuous relationships.

### Scatter Plots: `scatter()` üí°

Scatter plots display individual data points, typically showing the relationship between two numerical variables. They are excellent for identifying correlations or clusters.


```python
print("\n--- Scatter Plots ---")

# Data for scatter plot
np.random.seed(42) # for reproducibility
study_hours = np.random.rand(50) * 10
exam_scores = 50 + (study_hours * 5) + np.random.normal(0, 8, 50)
exam_scores = np.clip(exam_scores, 0, 100) # Clip scores to be between 0 and 100

plt.figure(figsize=(8, 6))
plt.scatter(study_hours, exam_scores,
            s=study_hours*20, # Marker size proportional to study hours
            c=exam_scores,    # Color based on exam score
            cmap='viridis',   # Colormap
            alpha=0.7,        # Transparency
            edgecolors='black') # Border color for markers

plt.title('Study Hours vs. Exam Scores')
plt.xlabel('Study Hours per Week')
plt.ylabel('Exam Score (%)')
plt.colorbar(label='Exam Score (%)') # Add a color bar
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```


**Key Use Cases:** Showing relationships/correlations between two variables, identifying clusters or outliers.

### Bar Plots: `bar()`, `barh()` üìè

Bar plots are used to compare quantities of different categories.

  * **`plt.bar(x, height)`:** Vertical bars. `x` typically represents categories, `height` their values.
  * **`plt.barh(y, width)`:** Horizontal bars.

<!-- end list -->


```python
print("\n--- Bar Plots ---")

# Data for bar plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 12, 39]

# Vertical Bar Plot
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'plum'])
plt.title('Sales by Category (Vertical Bar)')
plt.xlabel('Category')
plt.ylabel('Sales Volume')
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(8, 5))
plt.barh(categories, values, color='teal') # barh takes y-labels first, then widths
plt.title('Sales by Category (Horizontal Bar)')
plt.xlabel('Sales Volume')
plt.ylabel('Category')
plt.gca().invert_yaxis() # Invert y-axis to have category A at the top
plt.show()
```



**Key Use Cases:** Comparing discrete categories, showing counts or aggregated values.

### Histograms: `hist()` (Understanding Distributions) üìä

Histograms display the distribution of a single numerical variable. They divide the data into "bins" and show how many data points fall into each bin.

```python
print("\n--- Histograms ---")

# Data for histogram (e.g., student ages)
student_ages = np.random.normal(loc=22, scale=3, size=500) # Mean 22, Std Dev 3
student_ages = np.clip(student_ages, 18, 30) # Clip to a reasonable range

plt.figure(figsize=(8, 6))
# bins: number of bins or sequence of bin edges
# edgecolor: color of the bin borders
plt.hist(student_ages, bins=10, edgecolor='black', alpha=0.7, color='lightsalmon')

plt.title('Distribution of Student Ages')
plt.xlabel('Age')
plt.ylabel('Number of Students')
plt.show()

# Example with more bins and normalized
data_scores = np.random.randint(0, 100, 1000) # 1000 random scores between 0 and 99
plt.figure(figsize=(8, 6))
plt.hist(data_scores, bins=20, density=True, edgecolor='blue', alpha=0.6, color='lightblue')
plt.title('Distribution of Test Scores (Normalized)')
plt.xlabel('Score')
plt.ylabel('Probability Density')
plt.show()
```


**Key Use Cases:** Showing the shape of a distribution, identifying central tendency, spread, and skewness, detecting outliers.

  * `density=True`: Normalizes the histogram so that the sum of the bar areas (or height if bin width is 1) is 1. Useful for comparing distributions.

### Pie Charts: `pie()` ü•ß

Pie charts show the proportion of categories within a whole. They are best for a small number of categories (typically 2-5).


```python
print("\n--- Pie Charts ---")

# Data for pie chart
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10] # Percentages or absolute values
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0.1, 0, 0) # Explode the 2nd slice (Hogs)

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Favorite Animal Proportions')
plt.show()
```


**Key Use Cases:** Showing parts of a whole, proportions.

  * `autopct='%1.1f%%'`: Formats the percentage values displayed on slices.
  * `explode`: A tuple that specifies the fraction of the radius with which to offset each slice.
  * `startangle`: Rotates the start of the first slice.
  * `plt.axis('equal')`: Ensures the pie chart is circular.

### Box Plots: `boxplot()` (Understanding Data Spread and Outliers) üì¶

Box plots (or box-and-whisker plots) are excellent for visualizing the distribution of numerical data and detecting outliers across different groups.


```python
print("\n--- Box Plots ---")

# Data for box plot (simulating scores for different groups)
group1_scores = np.random.normal(70, 10, 100) # Mean 70, Std 10
group2_scores = np.random.normal(80, 5, 100)  # Mean 80, Std 5
group3_scores = np.random.normal(75, 12, 100) # Mean 75, Std 12
group3_scores = np.append(group3_scores, [30, 105]) # Add some outliers

data_to_plot = [group1_scores, group2_scores, group3_scores]
labels = ['Group A', 'Group B', 'Group C']

plt.figure(figsize=(10, 6))
plt.boxplot(data_to_plot, labels=labels, patch_artist=True, # Fill boxes with color
            boxprops=dict(facecolor='lightblue', medianprops=dict(color='red')))

plt.title('Exam Score Distribution by Group')
plt.xlabel('Group')
plt.ylabel('Scores')
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```


**Key Use Cases:** Comparing distributions between groups, identifying median, quartiles (spread), and potential outliers.

  * **Box:** Represents the interquartile range (IQR) ‚Äì the middle 50% of the data. The line inside is the median.
  * **Whiskers:** Extend from the box to indicate the variability outside the upper and lower quartiles, usually to 1.5 times the IQR.
  * **Outliers:** Points beyond the whiskers are considered outliers and are plotted individually.
  * `patch_artist=True`: Allows filling the boxes with color.
  * `boxprops`, `medianprops`: Dictionaries to customize the box and median line.

# 5.4 Customization and Annotation

Beyond just choosing the right plot type, effective data visualization requires careful customization and annotation to ensure clarity and convey your message.

### Setting Axis Limits (`xlim()`, `ylim()`) üîç

You can explicitly control the range of values displayed on the x and y axes. This is useful for focusing on specific data ranges or ensuring consistency across multiple plots.


```python
print("\n--- Setting Axis Limits ---")

x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c, s = np.cos(x), np.sin(x)

plt.figure(figsize=(10, 5))
plt.plot(x, c, color='blue', label='Cosine')
plt.plot(x, s, color='red', label='Sine')

# Set x-axis limits
plt.xlim(-2.5, 2.5) # From -2.5 to 2.5
# Set y-axis limits
plt.ylim(-1.2, 1.2) # From -1.2 to 1.2

plt.title('Cosine and Sine Waves with Custom Axis Limits')
plt.xlabel('Angle (radians)')
plt.ylabel('Value')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()

# You can also set limits on Axes object: ax.set_xlim(), ax.set_ylim()
```


**Code Explanation:**

  * `plt.xlim(min_val, max_val)`: Sets the lower and upper bounds of the x-axis.
  * `plt.ylim(min_val, max_val)`: Sets the lower and upper bounds of the y-axis.

### Adding Text and Annotations (`text()`, `annotate()`) üìù

Adding text directly to your plot can highlight specific points or add context.

  * **`plt.text(x, y, string, **kwargs)`:** Adds text at a specific data coordinate `(x, y)`.
  * **`plt.annotate(text, xy, xytext, arrowprops, **kwargs)`:** Adds text with an optional arrow pointing from the annotation text to a specific data point.

<!-- end list -->


```python
print("\n--- Adding Text and Annotations ---")

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
revenue = [100, 110, 95, 130, 120, 150]

plt.figure(figsize=(9, 6))
plt.plot(months, revenue, marker='o', color='purple', linestyle='-')

# Add a text label directly on the plot
plt.text(2.5, 100, 'Mid-quarter dip', fontsize=10, color='gray', ha='center')

# Annotate a specific point (e.g., peak revenue)
peak_month_idx = np.argmax(revenue)
peak_month = months[peak_month_idx]
peak_revenue = revenue[peak_month_idx]

plt.annotate(f'Peak Revenue: ${peak_revenue}k',
             xy=(peak_month_idx, peak_revenue), # Point to annotate
             xytext=(peak_month_idx + 0.5, peak_revenue + 10), # Text position
             arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8), # Arrow properties
             fontsize=10, color='darkgreen',
             bbox=dict(boxstyle="round,pad=0.3", fc="yellow", ec="blue", lw=0.5, alpha=0.8)) # Text box properties

plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue ($k)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```


**Code Explanation:**

  * `plt.text(x, y, string)`: Simple text at `(x, y)` data coordinates. `ha='center'` (horizontal alignment) centers the text.
  * `plt.annotate()`: More powerful for linking text to data points.
      * `xy`: The `(x, y)` coordinate of the point being annotated.
      * `xytext`: The `(x, y)` coordinate where the annotation text itself will appear.
      * `arrowprops`: A dictionary to customize the arrow connecting `xy` and `xytext`.
      * `bbox`: A dictionary to customize the text box around the annotation.

### Grid Lines (`grid()`) üï∏Ô∏è


**Code Explanation:**

  * `plt.grid(True)`: Turns the grid on.
  * `axis`: Specifies which axis (x, y, or both) should have grid lines.
  * `linestyle`, `linewidth`, `color`, `alpha`: Allow fine-tuning the appearance of the grid lines.

### Multiple Subplots (`subplot()`, `subplots()`) üñºÔ∏è

You can arrange multiple plots within a single Figure. This is critical for comparing different aspects of your data side-by-side.

  * **`plt.subplot(nrows, ncols, index)`:** Old-style, low-level function. Creates an Axes object at a specific position.
  * **`plt.subplots(nrows, ncols, **kwargs)`:** The recommended way (OO approach). Returns a `Figure` object and an array of `Axes` objects.

<!-- end list -->


```python
print("\n--- Multiple Subplots ---")

# Data for subplots
x_data = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x_data)
y_cos = np.cos(x_data)
y_tan = np.tan(x_data / 4) # Avoids vertical asymptotes for demo

# Method 1: Using plt.subplots() (Recommended)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8)) # 2x2 grid

# Plot 1: Top-Left
axes[0, 0].plot(x_data, y_sin, color='blue')
axes[0, 0].set_title('Sine Wave')
axes[0, 0].set_xlabel('X')
axes[0, 0].set_ylabel('Y')
axes[0, 0].grid(True)

# Plot 2: Top-Right
axes[0, 1].plot(x_data, y_cos, color='red')
axes[0, 1].set_title('Cosine Wave')
axes[0, 1].set_xlabel('X')
axes[0, 1].set_ylabel('Y')
axes[0, 1].grid(True)

# Plot 3: Bottom-Left (Example: Scatter Plot)
np.random.seed(0)
scatter_x = np.random.rand(50) * 10
scatter_y = np.random.rand(50) * 10
axes[1, 0].scatter(scatter_x, scatter_y, color='green', alpha=0.6)
axes[1, 0].set_title('Random Scatter')
axes[1, 0].set_xlabel('Var 1')
axes[1, 0].set_ylabel('Var 2')
axes[1, 0].grid(True)

# Plot 4: Bottom-Right (Example: Bar Plot)
categories = ['P1', 'P2', 'P3']
values = [15, 25, 10]
axes[1, 1].bar(categories, values, color=['cyan', 'magenta', 'orange'])
axes[1, 1].set_title('Category Counts')
axes[1, 1].set_xlabel('Category')
axes[1, 1].set_ylabel('Count')
axes[1, 1].grid(True)

plt.tight_layout() # Adjust subplot parameters for a tight layout
plt.suptitle('Various Subplots in a Single Figure', y=1.02, fontsize=16) # Super title for the whole figure
plt.show()


# Method 2: Using plt.subplot() (Less common for new code)
plt.figure(figsize=(10, 4)) # New figure for this example

plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
plt.plot(x_data, y_sin)
plt.title('Sine')

plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
plt.plot(x_data, y_cos)
plt.title('Cosine')

plt.suptitle('Subplots with plt.subplot()') # Super title for the whole figure
plt.tight_layout()
plt.show()
```


**Code Explanation:**

  * **`fig, axes = plt.subplots(nrows, ncols)`:** Creates a figure (`fig`) and an array of `Axes` objects (`axes`). If `nrows=2, ncols=2`, then `axes` will be a 2x2 NumPy array of `Axes` objects, which you access like `axes[row_index, col_index]`. If `nrows=1` or `ncols=1`, `axes` might be a 1D array.
  * **`axes[row, col].plot(...)`:** You call plotting methods directly on the specific `Axes` object.
  * **`plt.tight_layout()`:** Automatically adjusts subplot parameters for a tight layout, preventing titles/labels from overlapping.
  * **`plt.suptitle()`:** Adds a main title to the entire figure, distinct from individual subplot titles.
  * **`plt.subplot(nrows, ncols, index)`:** The older way. `index` starts from 1. You repeatedly call `plt.subplot` to select the "current" subplot before plotting with `plt.plot()`.




### Saving Plots (`savefig()`) üíæ

Once you've created a beautiful plot, you'll want to save it to a file.


### Displaying Images (`imshow()`) üì∏

`imshow()` is used to display data as an image, i.e., on a 2D regular grid. It's often used for displaying actual image files, heatmaps, or matrices as images.

For an image, `imshow()` expects a NumPy array.

  * **Grayscale images:** A 2D array (height x width).
  * **Color images:** A 3D array (height x width x 3 for RGB, or height x width x 4 for RGBA).

<!-- end list -->


```python
print("--- Displaying Images (imshow()) ---")

# 1. Creating and displaying a simple grayscale "image" (NumPy array)
# 10x10 array with a gradient
grayscale_data = np.arange(100).reshape((10, 10))

plt.figure(figsize=(6, 5))
plt.imshow(grayscale_data, cmap='gray') # 'gray' colormap for grayscale
plt.title('Grayscale Image (NumPy Array)')
plt.colorbar(label='Pixel Value')
plt.show()

# 2. Creating and displaying a simple color "image" (RGB array)
# A red square (20x20 pixels)
color_data = np.zeros((20, 20, 3), dtype=np.uint8) # Create a black image
color_data[5:15, 5:15, 0] = 255 # Set red channel in a square to max (255)
# Add some blue for a purplish effect outside the square
color_data[:, :, 2] = 100

plt.figure(figsize=(6, 5))
plt.imshow(color_data) # Matplotlib infers it's RGB
plt.title('Color Image (RGB NumPy Array)')
plt.axis('off') # Turn off axis ticks and labels for image
plt.show()

# 3. Example of a heatmap (often using a colormap)
heatmap_data = np.random.rand(8, 8) * 100 # Random data for a heatmap

plt.figure(figsize=(7, 6))
plt.imshow(heatmap_data, cmap='hot', interpolation='nearest') # 'hot' colormap, no interpolation for sharp pixels
plt.title('Heatmap Example')
plt.colorbar(label='Intensity')
plt.show()
```


**Code Explanation:**

  * `plt.imshow(data, cmap=..., interpolation=...)`: Displays the 2D (or 3D for color) NumPy array as an image.
      * `cmap`: Specifies the colormap to use (e.g., 'gray', 'hot', 'viridis'). More on colormaps later.
      * `interpolation`: Controls how pixels are rendered when zoomed (e.g., 'nearest' for sharp, 'bilinear' for smooth).
  * `plt.colorbar()`: Adds a color bar next to the image, showing the mapping from pixel values to colors.
  * `plt.axis('off')`: Hides the axes, which is common for displaying images where the coordinates aren't relevant.



### Basic 3D Plots (`mplot3d` toolkit) üßä

Matplotlib includes the `mplot3d` toolkit for creating 3D plots. You need to import `Axes3D` from `mpl_toolkits.mplot3d` for this functionality.

Common 3D plot types:

  * **3D Scatter Plot:** `ax.scatter(x, y, z)`
  * **3D Line Plot:** `ax.plot(x, y, z)`
  * **3D Surface Plot:** `ax.plot_surface(X, Y, Z)` (requires 2D arrays X, Y, Z)
  * **3D Wireframe Plot:** `ax.plot_wireframe(X, Y, Z)`

<!-- end list -->


```python
print("\n--- Basic 3D Plots (`mplot3d` toolkit) ---")

# 1. 3D Scatter Plot
fig = plt.figure(figsize=(8, 7))
ax = fig.add_subplot(111, projection='3d') # Crucial: specify projection='3d'

# Data for 3D scatter
x_scatter = np.random.rand(100) * 10
y_scatter = np.random.rand(100) * 10
z_scatter = np.random.rand(100) * 10

ax.scatter(x_scatter, y_scatter, z_scatter, c='red', marker='o', s=50, alpha=0.7)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Basic 3D Scatter Plot')
plt.show()


# 2. 3D Surface Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Data for a 3D surface (e.g., a "Mexican hat" function)
x_surf = np.linspace(-5, 5, 50)
y_surf = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x_surf, y_surf)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R) / R # A common surface function

ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.8) # viridis colormap
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot')
plt.show()
```


**Code Explanation:**

  * `fig.add_subplot(111, projection='3d')`: This is the key to creating a 3D Axes. `111` means 1x1 grid, 1st subplot.
  * `ax.scatter(x, y, z, ...)`: Plots points in 3D space.
  * `ax.plot_surface(X, Y, Z, ...)`: Creates a surface plot. `X`, `Y`, and `Z` must be 2D arrays, typically generated using `np.meshgrid()` from 1D arrays to cover the grid.
  * `cmap`: Colormap is often used to add another dimension of information (color mapping to Z values) on surfaces.

**Limitations:** While Matplotlib can do basic 3D plots, for highly interactive or advanced 3D visualizations, libraries like Plotly or Mayavi are often preferred.

### Customizing Ticks and Tick Labels üìä

Beyond just setting axis labels, you can control the positions and formatting of the tick marks and their corresponding labels.


```python
print("\n--- Customizing Ticks and Tick Labels ---")

x = np.linspace(0, 10, 50)
y = np.exp(-x / 2) * np.sin(x * 5)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y, color='darkgreen')

ax.set_title('Custom Ticks and Labels')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Amplitude (normalized)')

# 1. Setting specific tick locations
# Major ticks every 2 units on X-axis
ax.set_xticks(np.arange(0, 11, 2))
# Minor ticks every 0.5 units on X-axis
ax.set_xticks(np.arange(0, 10.5, 0.5), minor=True)

# Major ticks at specific values on Y-axis
ax.set_yticks([-0.5, 0, 0.5, 1.0])
# Minor ticks more frequently
ax.set_yticks(np.arange(-0.5, 1.01, 0.1), minor=True)

# 2. Customizing tick labels (e.g., adding units)
# Use a formatter for y-axis to add '%'
from matplotlib.ticker import PercentFormatter, FormatStrFormatter

# Format major y-ticks as percentage (multiplies by 100)
# ax.yaxis.set_major_formatter(PercentFormatter(xmax=1)) # xmax=1 because our data is already 0-1
# Or just a string formatter
ax.yaxis.set_major_formatter(FormatStrFormatter('%.1f')) # One decimal place

# Custom x-tick labels for certain points
custom_x_labels = [f'{i}s' for i in ax.get_xticks()] # Get current ticks, then format
ax.set_xticklabels(custom_x_labels)


ax.grid(True, which='major', linestyle='--', alpha=0.7)
ax.grid(True, which='minor', linestyle=':', alpha=0.3)

plt.show()
```


**Code Explanation:**

  * `ax.set_xticks(locations, minor=False/True)`: Sets the locations of major or minor ticks on the x-axis. Similar for `ax.set_yticks()`.
  * `ax.set_xticklabels(labels)`: Sets custom text labels for the ticks. The number of labels must match the number of ticks.
  * **Formatters (`matplotlib.ticker`):** Powerful tools to automatically format tick labels based on numerical values.
      * `PercentFormatter`: Formats as percentages.
      * `FormatStrFormatter`: Formats using a C-style format string (e.g., `%.1f` for one decimal place float).
  * `ax.xaxis.set_major_formatter()` / `ax.yaxis.set_major_formatter()`: Apply the formatter to the major ticks of an axis.

### Color Maps (`cmap`) üé®

Colormaps are used to map data values to colors, providing an additional visual dimension, especially in heatmaps, surface plots, and scatter plots where `c` (color) is mapped to a variable.


```python
print("\n--- Color Maps (cmap) ---")

# Data for a scatter plot with a third dimension mapped to color
np.random.seed(1)
x_data = np.random.rand(100) * 10
y_data = np.random.rand(100) * 10
z_data = np.random.rand(100) * 50 # Values to map to color

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot with 'viridis' colormap (perceptually uniform)
scatter1 = ax1.scatter(x_data, y_data, c=z_data, cmap='viridis', s=100, alpha=0.8, edgecolors='black')
ax1.set_title('Colormap: Viridis (Perceptually Uniform)')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
fig.colorbar(scatter1, ax=ax1, label='Z Value')

# Plot with 'hot' colormap (often used for heat)
scatter2 = ax2.scatter(x_data, y_data, c=z_data, cmap='hot', s=100, alpha=0.8, edgecolors='black')
ax2.set_title('Colormap: Hot (Sequential)')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
fig.colorbar(scatter2, ax=ax2, label='Z Value')

plt.tight_layout()
plt.show()

# You can find all colormaps here: https://matplotlib.org/stable/users/explain/colors/colormaps.html
```


**Code Explanation:**

  * `cmap='viridis'`: One of the "perceptually uniform" colormaps, recommended for their ability to accurately represent data changes across the entire range (e.g., 'viridis', 'plasma', 'cividis', 'magma', 'inferno').
  * `cmap='hot'`: A common "sequential" colormap, often used for heatmaps.
  * `fig.colorbar(mappable, ax=ax, label=...)`: Adds a color bar. `mappable` is the object that has a colormap (e.g., `scatter` object from `ax.scatter()`).



### Twin Axes (Multiple Y-axes) üëØ

Sometimes you need to plot two different datasets that share the same x-axis but have vastly different scales on their y-axes. Twin axes allow you to do this gracefully.


```python
print("\n--- Twin Axes (Multiple Y-axes) ---")

# Data for twin axes
time = np.linspace(0, 10, 100)
# Temperature in Celsius
temperature_c = 25 + 5 * np.sin(time)
# Pressure in kPa (higher values)
pressure_kpa = 101.3 + 5 * np.cos(time * 0.5) * 5 # Random variations around standard pressure

fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot temperature on the first y-axis (ax1)
ax1.plot(time, temperature_c, color='red', label='Temperature (¬∞C)')
ax1.set_xlabel('Time (hours)')
ax1.set_ylabel('Temperature (¬∞C)', color='red')
ax1.tick_params(axis='y', labelcolor='red') # Make tick labels match line color
ax1.set_title('Temperature and Pressure over Time')
ax1.grid(True, linestyle='--', alpha=0.6)


# Create a second y-axis that shares the same x-axis as ax1
ax2 = ax1.twinx()

# Plot pressure on the second y-axis (ax2)
ax2.plot(time, pressure_kpa, color='blue', linestyle='--', label='Pressure (kPa)')
ax2.set_ylabel('Pressure (kPa)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Combine legends from both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

plt.show()
```


**Code Explanation:**

  * `ax2 = ax1.twinx()`: This is the magic line. It creates a new `Axes` object (`ax2`) that has an invisible x-axis overlaid on `ax1`'s x-axis, but a completely independent y-axis.
  * `ax1.tick_params(axis='y', labelcolor='red')`: Customize tick parameters for a specific axis.
  * **Combining legends:** When using twin axes, you need to manually collect the legend handles and labels from both axes and then call `legend()` on one of them (e.g., `ax2`) with the combined lists.



### Animations (`FuncAnimation`) - *brief introduction* üé¨

Matplotlib can create basic animations. This typically involves updating data points in a loop and redrawing the plot for each frame. The `FuncAnimation` class from `matplotlib.animation` is the primary tool.

### Integrating Matplotlib with NumPy and Pandas ü§ù

Matplotlib is designed to work seamlessly with NumPy arrays, which we've used extensively for data generation. It also integrates beautifully with Pandas DataFrames.

#### NumPy Integration (Already Demonstrated Heavily)

All plotting functions (e.g., `plt.plot()`, `plt.scatter()`, `plt.hist()`) accept NumPy arrays directly as input for their data. This is fundamental to Matplotlib's operation.

#### Pandas Integration

When your data is structured in a Pandas DataFrame, you can directly pass DataFrame columns or Series to Matplotlib plotting functions. Pandas itself also has built-in plotting capabilities (which internally use Matplotlib).

```python
print("\n--- Integrating Matplotlib with NumPy and Pandas ---")

# Data for Pandas DataFrame
data = {
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'New York', 'Los Angeles', 'Chicago'],
    'Temperature': [25, 30, 20, 32, 35, 27, 28, 22], # Celsius
    'Humidity': [70, 60, 80, 75, 40, 68, 62, 78], # Percent
    'Population': [8.4, 3.9, 2.7, 2.3, 1.6, 8.4, 3.9, 2.7] # Millions
}
df = pd.DataFrame(data)
print("Pandas DataFrame:")
print(df)

# Plotting directly from Pandas DataFrame columns
fig, ax = plt.subplots(figsize=(10, 6))

# Use DataFrame columns directly as x and y values
# pandas.DataFrame.plot() is a wrapper for matplotlib
df.plot(kind='scatter', x='Temperature', y='Humidity', s=df['Population']*50, # Marker size by population
        c='Population', cmap='coolwarm', alpha=0.7, ax=ax, edgecolors='black',
        title='City Temperature vs. Humidity (Population Size)')
ax.set_xlabel('Temperature (¬∞C)')
ax.set_ylabel('Humidity (%)')
fig.colorbar(ax.collections[0], ax=ax, label='Population (Millions)') # Access scatter collection for colorbar

plt.show()

# Example: Bar plot of average temperature by city using Pandas group-by and plot
avg_temp_by_city = df.groupby('City')['Temperature'].mean()
print("\nAverage Temperature by City (Pandas Series):")
print(avg_temp_by_city)

plt.figure(figsize=(8, 5))
avg_temp_by_city.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Average Temperature by City')
plt.xlabel('City')
plt.ylabel('Average Temperature (¬∞C)')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for readability
plt.tight_layout()
plt.show()
```


**Code Explanation:**

  * `df.plot(kind='scatter', x='Temperature', y='Humidity', ...)`: Pandas DataFrames have a `.plot()` method that provides a convenient interface to Matplotlib plotting functions. You specify the `kind` of plot ('line', 'bar', 'hist', 'scatter', 'pie', 'box', etc.) and map DataFrame columns to x and y axes.
  * `ax=ax`: Crucially, by passing an `Axes` object (`ax`), you ensure that the Pandas plot is drawn onto your pre-defined Matplotlib `Axes` rather than creating a new one. This allows for fine-tuning with `ax` methods.
  * `df.groupby('City')['Temperature'].mean().plot(kind='bar', ...)`: Demonstrates chaining Pandas operations (grouping and aggregation) with its built-in plotting capability.
  * `plt.xticks(rotation=45, ha='right')`: Important for bar charts with long category names to prevent labels from overlapping.

**Key Takeaway:** Matplotlib's strength lies in its deep integration with the data structures of NumPy and Pandas. This synergy forms the backbone of data analysis and visualization workflows in Python.
